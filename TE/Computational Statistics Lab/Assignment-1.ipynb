```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('real_estate_price_size_year_view.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
print("Price")
print("Mean: ", df['price'].mean())
print("Median: ", df['price'].median())
print("Mode: ", df.mode()['price'][0])
print()

print("Size")
print("Mean: ", df['size'].mean())
print("Median: ", df['size'].median())
print("Mode: ", df.mode()['size'][0])
```

```python
correlation_matrix = df[['price', 'size', 'year']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(6, 4))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Line Graph

```python
avg_price_by_year = df.groupby('year')['price'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.plot(avg_price_by_year['year'], avg_price_by_year['price'], color='red', marker='o')
plt.title('Price vs. Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

### Bar Graph

```python
plt.figure(figsize=(8, 6))
plt.bar(avg_price_by_year['year'], avg_price_by_year['price'])
plt.title('Average Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

```python
view_counts = df['view'].value_counts()

plt.figure(figsize=(8, 6))
view_counts.plot(kind='bar')
plt.xlabel('View Type')
plt.ylabel('Count')
plt.title('Properties per View Type')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

### Scatter Plot

```python
# using plt.scatter()

plt.figure(figsize=(8, 6))
plt.scatter(df['size'], df['price'])
plt.xlabel('Size')
plt.ylabel('Price')
plt.title('Price vs. Size')
plt.grid(True)
plt.show()
```

```python
# using sns.scatterplot()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='size', y='price', data=df, hue='view', palette={'Sea view': 'blue', 'No sea view': 'green'})
plt.xlabel('Size')
plt.ylabel('Price')
plt.title('Price vs. Size (by View Type)')
plt.grid(True)
plt.show()
```

### Histogram

```python
plt.figure(figsize=(8, 6))
plt.hist(df['price'], bins=20, color='lightpink', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['size'], bins=20, color='purple', edgecolor='black')

plt.xlabel('Size')
plt.ylabel('Frequency')
plt.title('Size Distribution')
plt.grid(axis='y')
plt.show()
```
