### Tokenization

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentence_tokenized = sent_tokenize(text)
print(sentence_tokenized)
```

```python
# tokenized = word_tokenize(text)
# print(tokenized)
```

```python
import string

def tokenization(sentence):
    punctuation = string.punctuation + '[]{}()<>'  # add brackets to punctuation set
    
    # for every punctuation mark found in the sentence, replace all instances
    for char in punctuation:
        sentence = sentence.replace(char, '')
        
    sentence = sentence.lower()
    tokens = sentence.split()
    
    return tokens
```

```python
tokens = []
for i in sentence_tokenized:
    tokens.extend(tokenization(i))
    
print(tokens)
```

### Stop Words Removal

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
filtered = []
for t in tokens:
    if t not in stop_words:
        filtered.append(t)
        
print(filtered)
```

### Stemming

```python
stemmer = PorterStemmer()
for t in tokens:
    print(f'{t} : {stemmer.stem(t)}')
```

### Term Frequency and Inverse Document Frequency

```python
def calculateTF(tokens):
    tf = {}
    for t in tokens:
        if t not in tf:
            tf[t] = tokens.count(t) / len(tokens)
            
    return tf
```

```python
calculateTF(tokens)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# documents
d0 = sentence_tokenized[0]
d1 = sentence_tokenized[1]
d2 = sentence_tokenized[2]

corpus = [d0, d1, d2]

tfidf = TfidfVectorizer()
```

```python
print('Word Indexes')
print(tfidf.vocabulary_)
```

```python
result = tfidf.fit_transform(corpus)
print('TF-IDF Value')
print(result)
```
