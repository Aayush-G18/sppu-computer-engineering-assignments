## Part II
Using iris.csv, a dataset that includes features of 3 different species of the iris flower.

```python
data = pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
numcols = [col for col in data.columns if (data[col].dtype=="float" or data[col].dtype=="int")]
numcols
```

```python
numeric = data._get_numeric_data()
numeric
```

### Mean

```python
for i in numcols:
    print(i, ": ", data[i].mean())
```

```python
for i in numcols:
    sum = 0
    for j in range(150):
        sum += data[i][j]
    print(i ,": ", sum/150)
```

### Median

```python

for i in numcols:
    print(i ,": ", data[i].median())
```

```python
for i in numcols:
    sorted_list = sorted(list(data[i]))
    print(i,": ", sorted_list[len(sorted_list)//2])
```

### Mode

```python
for i in range(5):
    print(data.columns[i], ": ", data.mode().values[0][i])
```

### Percentile

```python
for i in numcols:
    tile = np.percentile(data[i], 25)
    print(i, "25 tile : " ,tile)
    tile = np.percentile(data[i], 50)
    print(i, "50 tile : " ,tile)
    tile = np.percentile(data[i], 75)
    print(i, "75 tile : " ,tile)
    print()
```

### Variance

```python
for i in numcols:
    variance = np.var(data[i])
    print(i, ": ",variance)
```

### Standard Deviation

```python
numeric.std()
```

```python
for i in numcols:
    sd = np.std(data[i])
    print(i,": ",sd)
```

### Grouping

```python
grouped_data = data.groupby("species")
```

```python
grouped_data.mean()
```

```python
grouped_data.median()
```

```python
grouped_data.min()
```

```python
grouped_data.max()
```

```python
grouped_data.std()
```
