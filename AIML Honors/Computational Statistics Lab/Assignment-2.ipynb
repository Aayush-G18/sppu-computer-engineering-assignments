```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TE_CSV.csv')
df.head()
```

```python
df.drop(df.iloc[:, 2:], inplace=True, axis=1)
df.head(15)
```

```python
df.describe()
```

```python
print('Mean = ', df['Marks'].mean())
print('Median = ', df['Marks'].median())
print('Std. Deviation = ', df['Marks'].std())
```

```python
correlation_matrix = df[['RollNo', 'Marks']].corr()
print(correlation_matrix)
```

### Line Graph

A line graph is not exactly useful in this case since it is used for continuous values and marks are discrete, whereas roll numbers are categorical. It may be useful if we want to visualize a single student's marks over time.

```python
plt.figure(figsize=(8, 6))
plt.plot(df['RollNo'], df['Marks'], marker='o')
plt.xlabel('Roll Number')
plt.ylabel('Marks')
plt.title('Marks vs. Roll Number')
plt.grid(True)
plt.show()
```

### Bar Graph

```python
bins = [0, 49, 69, 79, 89, 100]
labels = ['F', 'D', 'C', 'B', 'A']

# grouping marks
marks_bins = pd.cut(df['Marks'], bins=bins, labels=labels, right=False)

# calculating no. of students in each bin
grade_counts = marks_bins.value_counts().sort_index()

plt.figure(figsize=(8, 6))
plt.bar(grade_counts.index, grade_counts.values)
plt.xlabel('Grade')
plt.ylabel('No. of Students')
plt.title('Grade Distribution')
plt.grid(axis='y')
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['RollNo'], df['Marks'])
plt.xlabel('Roll Number')
plt.ylabel('Marks')
plt.title('Marks vs. Roll Number')
plt.grid(True)
plt.show()
```

### Histogram

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Marks'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Distribution')
plt.grid(axis='y')
plt.show()
```
